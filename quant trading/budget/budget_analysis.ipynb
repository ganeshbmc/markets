```python
df["budget_date"] = pd.to_datetime(df["budget_date"], format="%d/%m/%Y")
df["peak_date"] = pd.to_datetime(df["peak_date"], format="%d/%m/%Y")
df["post_date"] = pd.to_datetime(df["post_date"], format="%d/%m/%Y")
df.dtypes
```

```python
df["peak_days"] = df["budget_date"] - df["peak_date"]
df["iv_drop_peak_to_bday"] = df["iv_peak"] - df["iv_b_day"]
df["nifty_change_peakday_bday"] = (df["nifty_spot_b_day"] - df["nifty_spot_iv_peak"])*100 / df["nifty_spot_iv_peak"]
df
```

```python
df["peak_post_days"] = df["post_date"] - df["peak_date"]
df["iv_drop_peak_to_post"] = df["iv_peak"] - df["iv_post"]
df["nifty_change_peakday_postday"] = (df["nifty_spot_iv_post"] - df["nifty_spot_iv_peak"])*100 / df["nifty_spot_iv_peak"]
df
```
