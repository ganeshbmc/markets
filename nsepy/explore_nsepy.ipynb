```python
import nsepy
```

```python
from nsepy import get_history
from datetime import datetime
```

`Getting stock data`  

```python
symbol = 'SBIN'

start = datetime(2022, 1, 1)
end = datetime.now()

sbin = get_history(symbol=symbol, 
                    start=start, 
                    end=end,)
sbin.tail()
```

`Getting F&O data`  

```python
exp = datetime(2022,8,25)

nifty_fut = get_history(symbol="NIFTY",
                        start=exp,
                        end=exp,
                        index=True,
                        futures=True,
                        expiry_date=exp)
nifty_fut.tail()
```

```python
exp = datetime(2022,8,25)

nifty_opt = get_history(symbol="NIFTY",
                        start=exp,
                        end=exp,
                        index=True,
                        option_type='CE',
                        strike_price=18000,
                        expiry_date=exp)
nifty_opt.tail()
```
